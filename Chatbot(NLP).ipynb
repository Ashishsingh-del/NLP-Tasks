```python
# Install sumy
!pip install sumy

# Import nltk and download required resources
import nltk
nltk.download("punkt")
nltk.download("punkt_tab")  # sometimes needed in Colab

# Import sumy modules
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lex_rank import LexRankSummarizer

# Sample news article text
news_article = """
The government announced a new economic policy today aimed at boosting exports
and supporting small businesses. The finance minister said the policy includes
tax reductions for startups and subsidies for exporters. Analysts predict this
move could increase the country's GDP growth by 1.5% over the next fiscal year.
Meanwhile, opposition leaders criticized the plan, arguing it favors large
corporations instead of addressing unemployment.
"""

# Create parser
parser = PlaintextParser.from_string(news_article, Tokenizer("english"))

# Use LexRank Summarizer
summarizer = LexRankSummarizer()

# Generate summary (top 2 sentences)
summary = summarizer(parser.document, 2)

print("Daily News Summary:")
for sentence in summary:
    print("-", sentence)
```
